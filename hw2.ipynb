```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
diabetes=datasets.load_diabetes()
```

```python
X = diabetes.data[:,2]
Y = diabetes.target
X_p=[X,Y]
plt.scatter(X_p[0],X_p[1])
plt.plot([0.15],[50],'r*')
plt.plot([0.15],[175],'k*')
plt.plot([0.15],[300],'b*')
plt.xlabel('Blood Pressure')
plt.ylabel('Diabetes')
```

```python
temp=[]
for (i,j) in zip(X_p[0],X_p[1]):
    r1=(i-0.15)**2 + (j-50)**2
    r2=(i-0.15)**2 + (j-175)**2
    r3=(i-0.15)**2 + (j-300)**2
    m=min(r1,r2,r3)
    temp.append([r1,r2,r3].index(m))
```

```python
x1=np.transpose([[i,j] for (t,i,j) in zip(temp,X_p[0],X_p[1]) if(t==0)])
x2=np.transpose([[i,j] for (t,i,j) in zip(temp,X_p[0],X_p[1]) if(t==1)])
x3=np.transpose([[i,j] for (t,i,j) in zip(temp,X_p[0],X_p[1]) if(t==2)])
r1=np.mean(x1[0]),np.mean(x1[1])
r2=np.mean(x2[0]),np.mean(x2[1])
r3=np.mean(x3[0]),np.mean(x3[1])
```

```python
plt.scatter(X_p[0],X_p[1],c=temp)
plt.plot(r1[0],r1[1],'r*')
plt.plot(r2[0],r2[1],'k*')
plt.plot(r3[0],r3[1],'b*')
plt.xlabel('Blood Pressure')
plt.ylabel('Diabetes')
```
